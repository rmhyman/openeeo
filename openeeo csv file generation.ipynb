```python
import csv


        
def write(w,d):
    w.writerow([d['year'],d['company'],d['category'],d['ethn'],d['gender'],d['count']])
    

def write_gender(w,d,gender,l,start):
    d['gender'] = gender
    i = start
    ethns = ['White','Black','Pacific Islander','Asian', 'American Indian','Two or More Races']
    for i in range(0,len(ethns)):
        d['ethn'] = ethns[i]
        d['count'] = l[start + i]
        write(w,d)

def write_hispanic(w,d,l,pos):
    d['ethn'] = 'Hispanic'
    d['count'] = l[pos+0]
    d['gender'] = 'Male'
    write(w,d)
    d['gender'] = 'Female'
    d['count'] = l[pos+1]
    write(w,d)



    

def write_category(w,d,l,pos):
    write_hispanic(w,d,l,pos)
    write_gender(w,d,'Male',l,pos+2)
    write_gender(w,d,'Female',l,pos+8)
    
def gen_csv(l,filename,year,company):
    header = ['Year','Company','Category','Ethnicity','Gender','Count']
    with open(filename,"wb") as csvfile:
        w = csv.writer(csvfile)
        w.writerow(header)
        max_ = len(l)
        i = 0
        d = {'year':year,
            'company': company,
            'category':'Executive/Sr. Officials/Mgrs' }
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'First/Mid Managers'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Professionals'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Technicians'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Sales Workers'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Admin Support'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Craft Workers'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Operatives'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Laborers & Helpers'
        write_category(w,d,l,i)
        i+=15
        d['category'] = 'Service Workers'
        write_category(w,d,l,i)
        i+=15
```

```python
gen_csv(l,"Microsoft/eeo1-2015.csv",'2015',"Microsoft")
```
